```python
import pandas as pd
import geopandas as gpd
```

```python
import fiona
fiona.supported_drivers  
```

```python
map_shape_file_path = '../data/map_of_ireland/IRL_adm1.shp'
map_dataset_path = 'output/1400_map_dataset.geojson'
```

```python
map_dataset = gpd.read_file(map_shape_file_path)
map_dataset = map_dataset.drop(columns=['ID_0','ISO','NAME_0','ID_1','TYPE_1','ENGTYPE_1','NL_NAME_1','VARNAME_1'])
map_dataset = map_dataset.rename(columns={'NAME_1':'County'})
map_dataset.plot()
```

```python
map_dataset['County'] = map_dataset['County'].apply(lambda x: x.replace('Laoighis', 'Laois'))
map_dataset.head()
```

```python
map_dataset.to_file(map_dataset_path)
print("Saved map dataset to: "+map_dataset_path)
```
