```python
import pandas as pd
import geopandas as gpd
```

```python
map_dataset_path = 'output/1400_map_dataset.geojson'
map_dataset_output_path = 'output/1410_map_dataset.geojson'
```

```python
map_dataset = gpd.read_file(map_dataset_path)
map_dataset.plot()
```

```python
map_dataset['geometry'] = map_dataset.simplify(0.005)
map_dataset.plot()
```

```python
map_dataset.to_file(map_dataset_output_path)
print("Saved map dataset to: "+map_dataset_output_path)
```
